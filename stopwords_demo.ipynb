```python
import stopwordsiso as stopwords

# We use "stopwordsiso" as a package name instead of "stopwords-iso",
# as package name in Python should not has an hyphen.
```

```python
stopwords.has_lang("th") # check if there is a stopwords list for the language
```

```python
stopwords.langs()  # return set of all the supported languages
```

```python
stopwords.stopwords("en")  # English stopwords
```

```python
stopwords.stopwords(["ja", "th", "en"])  # Japanese, Thai, and English stopwords
```

```python
stopwords.stopwords("xxx")  # if the language is not supported, an empty set will be returned
```
